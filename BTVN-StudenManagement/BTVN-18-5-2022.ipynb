```python
class Student:
    def __init__(self, code, fullname, dob, residence):
        self.code = code
        self.fullname = fullname
        self.dob = dob
        self.residence = residence
    def info(self):
        print('{0}\t\t{1}\t\t{2}\t\t{3}'.format(self.code,self.fullname,self.dob,self.residence))
        
```

```python
class ListStudent:
    def __init__(self, _list):
        self.list = _list
    def initialize(self):
        st1 = Student('001','Nguyen Van A', '01/01/1999','Ha Noi')
        st2 = Student('002','Nguyen Van B', '02/01/1999','Ha Noi')
        st3 = Student('003','Nguyen Van C', '03/01/1999','Ha Noi')
        st4 = Student('004','Nguyen Van D', '04/01/1999','Ha Noi')
        self.list = [st1,st2,st3,st4]
    def output(self):
        print('MSSV\t\tHo va Ten\t\tNgay sinh\t\tQue quan')
        for i in self.list:
            i.info()
    def add(self, std):
        self.list.append(std)
    def update(self, code, std):
        '''code:ma sinh vien duoc update/std: thong tin moi'''
        for i in self.list:
            if i.code == code:
                i.fullname = std.fullname
                i.dob = std.dob
                i.residence = std.residence
    def delete(self, code):
        for i in self.list:
            if i.code == code:
                self.list.remove(i)
                break
    def find(self, code):
        for i in self.list:
            if i.code == code:
                i.info()
                break
```

```python
class StudentManagement:
    list_st = None 
    def menu(self):
        print('=== STUDENTS LIST MANAGEMENT')
        print('= 0. Load list student')
        print('= 1. Show students menu')
        print('= 2. Add new student')
        print('= 3. Update students information')
        print('= 4. Delete student')
        print('= 5. Find')
        print('= 6. Quit')
        print('===============================')
    def run(self):
        while True:
            self.menu()
            action = input('Whats you wanna do?')
            if action == '0':
                self.list_st = ListStudent(None)
                self.list_st.initialize()
                print('Loading 100%')
                input('Press Enter to continue')
            elif action =='1':
                if self.list_st == None:
                    print('Still unload student yet')
                else:
                    self.list_st.output()
                input('Press Enter to continue')
            elif action == '2':
                code = input('Enter student id: ')
                fullname = input('Enter student full name: ')
                dob = input('Enter student date of bỉth: ')
                residence = input('Enter student residence: ')
                adding_student = Student(code, fullname, dob, residence)
                self.list_st.add(adding_student)
                print('Adding completely')
                input('Press enter to continue')
            elif action == '3':
                code = input('Enter student id: ')
                fullname = input('Enter student full name: ')
                dob = input('Enter student date of bỉth: ')
                residence = input('Enter student residence: ')
                update_student = Student(code, fullname, dob, residence)
                self.list_st.update(update_student.code,update_student)
                print('Updated completely')
            elif action == '4':
                code = input('Enter student id: ')
                self.list_st.delete(code)
                print('Completely delete')
            elif action == '5':
                code = input('Enter student id: ')
                self.list_st.find(code)
                print('Result here')
            elif action == '6':
                break
        print('End')
```

```python
sm = StudentManagement()
sm.run()
```
